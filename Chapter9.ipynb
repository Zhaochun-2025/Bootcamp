```python
# Monday: Generators and Iterators
# creating a basic iterator from an iterable
sports = ["baseball", "soccer", "football", "hockey", "basketball"]
my_iter = iter(sports)
print( next(my_iter)) 
print( next(my_iter))
for item in my_iter:
    print(item)
#print( next(my_iter)) # This will raise StopIteration since the iterator is exhausted

# Creating our own iterator
class Alphabet():
    def __iter__ (self):
        self.letters = "abcdefghijklmnopqrstuvwxyz"
        self.index = 0
        return self
    def __next__(self):
        if self.index <= 25:
            char = self.letters[ self.index ]
            self.index += 1
            return char
        else:
            raise StopIteration #stops the iteration once all letters are returned
# Using our custom iterator
for char in Alphabet():
    print(char)

# Creating a Range Generator
def myRange(stop, start = 0, step =1):
    while start < stop: 
        print( "Generator Start Value: {}" .format(start))
        yield start
        start += step # increment start, otherwise infinite loop
for x in myRange(5): 
    print( "For Loop X Value: {}" .format(x))

#HWK
#reverse iterate a list that returns information in a reverse order
def reverse_iterable(iterable):
    index = len(iterable) - 1
    while index >= 0:
        yield iterable[index]
        index -= 1
# Testing the reverse_iterable function
for i in reverse_iterable ([1, 2, 3, 4, 5]):
    print(i)

# 2
# Create a generator that acts like the range function but yields a squared number everytime
def squared_range(stop, start = 0, step = 1):
    while start < stop:
        yield start ** 2
        start += step
for i in squared_range(4):
    print(i)

# Tuesday: Decorators
# Decorators give other functions extra capabilities without modifying them by "@" denote

```
