```python
# Thursday: Recursive Functions and Memoization 
# Recursive- where a function calls itself one or more tiems within its block 
# Memoization- helps to speed up recursive functions, by storing results of expensive function calls and returning the cached result when the same inputs occur again

# Writing a Factorial Function
def factorial(n):
    # set your base case
    if n <= 1:
        return 1
    else:
        return factorial( n-1 ) * n
print( factorial(5))

# The Fibonacci Sequence: each number in the sequence is the sum of the previous two numbers
def fib(n):
    if n <= 1:
        return n 
    else: 
        return fib( n-1 ) + fib( n-2 )
print( fib(5) )

# Understanding Memoization
# Memoization is a technique to speed up recursive functions by storing previously computed results
cache = {} # used to cache values to be used later
def fib(n):
    if n in cache:
        return cache[ n ]
    result = 0
    # base case
    if n <= 1:
        result = n
    else: 
        result = fib( n-1 ) + fib( n-2 )
    cache[ n ] = result # save result into dictionary iwth n as the key 

    return result
print(fib(50))

#Using @lru_cache
from functools import lru_cache
@lru_cache()
def fib(n): 
    if n <= 1:
        return n 
    else:
        return fib( n-1 ) + fib( n-2 )
fib(50)

#hwk
#1
#apply lru_cache built-in decorator to the factorial function
from functools import lru_cache
@lru_cache()
def factorial(n):
    if n <= 1:
        return 1
    else:
        return factorial(n - 1) * n
print(factorial(5))

#2
#create function that takes in two arguments, a list and a number, and returns the index of the number in the list using recursion
def find_index(lst, item, index=0):
    if index >= len(lst):
 
```
