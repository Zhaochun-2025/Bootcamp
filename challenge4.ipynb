```python
import matplotlib.pyplot as plt #external library for plotting

x = [1,2,3,4,5]
y=[2,3,5,7,11]

# create a line plot using Matplotlib to visualize the data
plt.plot(x, y, marker='o', linestyle='-', color='b') # 'b' for blue color,'o' for circle markers, '-' for solid line
# add title and labels
plt.title('Sample Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# display the plot
plt.show()

# perform quadratic plotting using the list above
import numpy as np
x = np.linspace(0, 10, 100)  # generate 100 points between 0 and 10
y = x**2  # calculate y as the square of x
# create a quadratic plot
plt.plot(x, y, color='r')  # 'r' for red color
# add title and labels
plt.title('Quadratic Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# display the quadratic plot
plt.show()
```
