```python
#MIPS Computer Literacy Bootcamp: Coding project
# This code was written supported by OpenAI's GPT-3.5 model

# Install missing packages
%pip install scipy pillow

#import necessary libraries
from PIL import Image # Pillow for image processing (opening, resizing, saving images)
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter # Gaussian filter for blurring
from scipy import ndimage, stats 

# Convert to requirements ( 500x500 pixel image, 8bit/greyscale)
original_image = Image.open('cat_window.PNG').convert('L') # Convert to grayscale
original_image = original_image.resize((500, 500)) # Resize to 500x500
original_image = np.array(original_image) # Convert to numpy array to process and analyse image data

plt.imshow(original_image, cmap='gray') # Display the original image in grayscale
plt.title('Original image') # Title for the original image before blurring
plt.axis('off') # Hide axes
plt.show() # Display the original image

# Apply Gaussian blurring
blurred_image = ndimage.gaussian_filter(original_image, sigma=2) # Apply Gaussian filter with sigma=2 for blurring

plt.imshow(blurred_image, cmap='gray') # Display the blurred image in grayscale
plt.title('Blurred Image') # Title for the blurred image
plt.axis('off') # Hide axes
plt.show() # Display the blurred image

#Histogram of Original vs Blurred
plt.figure(figsize = (10, 5)) # Set figure size for the histogram to be produced

#Histogram for original image
hist_orig, bins_orig = np.histogram(original_image.flatten(), bins=256, range=[0, 256]) 
plt.plot(hist_orig, label='original', color='blue') #Sets the legend for the original image histogram

#Histogram for blurred image
hist_blur, bins_blur = np.histogram(blurred_image. flatten(), bins= 256, range= [0, 256])
plt.plot(hist_blur, label= 'blurred', color= 'red') #Sets the legend for the blurred image histogram

plt.legend() # Add legend to the histogram
plt.title('Histogram of Original vs Blurred Image')
plt.xlabel('Pixel Intensity') # X-axis label for pixel intensity
plt.ylabel('Frequency') # Label for the y-axis
plt.grid(True) # Add grid to the histogram
plt.show() # Display the histogram

# Statistical comparison using Kolomogorov-Smirnov test (from Scipy stats) for statistical different between histograms
ks_statistic, p_value = stats.ks_2samp(hist_orig, hist_blur)
print( "=== statistical analyis (Kolmogorov-Smirnov test) ===")
print(f"Kolmogorov-Smirnov Statistic: {ks_statistic}, P-value: {p_value}")
if p_value < 0.05:
    print("The two histograms are significantly different.")
else:
    print("The two histograms are not significantly different.")

# Save the blurred image
blurred_image = Image.fromarray(blurred_image) # Convert numpy array back to PIL image
blurred_image.save('blurred_cat_window.png') # Save the blurred image


# Downscale to 250 x 250
downscale_image = Image.fromarray(original_image).resize((250, 250))
downscale_image.save('downscaled_blurred_cat_window.png') # Save the downscaled image

plt.imshow(np.array(downscale_image), cmap='gray') # Display the downscaled image in grayscale
plt.title('Downscaled Image') # Title for the downscaled image
plt.axis('off') # Hide axes
plt.show() # Display the downscaled image

# Print confirmation of saved images
print( "Image saved: 'blurred_cat_window.png' and 'downscaled_blurred_cat_window.png'")



```
