```python
#Tuesday; Working with Dictionaries
#Adding New information
car = {"year": 2018}
car["color"] = "Blue"
print( "year:{} \t color:{}" .format( car["year"], car["color"]))

#Changing Information
# updating a value for a key/value pair that already exists
car = { "year": 2018, "color": "Blue" }
car ["color"] = "Red"
print( "Year: {} \t Color: {}" .format( car["year"], car["color"]))

#Deleting Information
car = {"year": 2018}
try:
    del car["year"]
    print(car)
except:     #incase crashing happens when key trying to be removed doesn't exist
    print("That key does not exist")

#Looping a Dictionary

#Looping Only Keys (ALL VARIABLES, without quotations)
person = {"name": "John", "age": 26}
for key in person.keys():
    print(key)
    print( person [key])    # will output the value at the current key
#Looping Only Values    (ONLY VALUES)
person = { "name": "John", "age": 26 }
for value in person.values():
    print(value)
#Looping Key-Value pairs (EVERYTHING)
person = { "name": "John", "age": 26}
for key, value in person.items():
    print( "{}:{}" .format(key, value))

#HWK
#1
user_info = { }
user_info["name"] = input("what is your name")
user_info["address"]= input("what is your address")
user_info["number"]= input("what is your number")
print("name: {}, address: {}, number: {}".format(
    user_info["name"], user_info["address"], user_info["number"]))

#2
person = { "name": "John Smith" }
print(person["name"])
```

```python
#Wednesday: Tupels, sets frozensets
# Tuple- like a list, but cannot be altered 
    #delcaring a tuple
t1 = ("hello", 2, "hello")  #with parens
t2 = True, 1        # without parens
print( type(t1), type(t2))      #both are tuples
    # t1[0] = 1   #will not work, as tuples are immutable 


# Set- can only contain unique values, unordered- cannot be accessed by index
# unordered- only accessible for the value itself
# set([]) OR {, , ,}
    #declaring a set
s1= set ([1, 2, 3, 1])  # use set keyword and square brackets
s2 = {4, 4, 5}      #uses curly brackets, like dictionary
print( type(s1), type(s2))
s1.add(5)
s1.remove(1)
print(s1)

#Frozensets
#combination of a set & tuple. They are immutable, unordered and unique
    #declaring a frozenset
fset = frozenset( [1, 2, 3, 4])
print( type(fset))

#hwk 
list = [ ]      #stored within list initially

while True:     #allow input as many numbers, but 'break' when 'done' is entered
    bank_number= input("enter desired bank account numbers, type 'done' when finished")
    if  bank_number.lower() == 'done':
        break
    list.append(bank_number)

list_set= frozenset(list)
print(list_set)

#2
nums = [3, 4, 3, 7, 10]
unique_values = frozenset(nums)
print( unique_values)


```
