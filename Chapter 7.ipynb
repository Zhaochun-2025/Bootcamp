```python
# Thursday: Inheritance
# Inheritance allows a class to inherit attributes and methods from another class 
        #from Superclass --> Subclass

# e.g., a Dog and Cat class will have the same code, attributes and methods

#Inheriting from a Class
class Animal():
    def makeSound(self):
        print("roar")
class Dog(Animal):  #INHERITING animal class
    species = "canine"

sam = Dog()
sam.makeSound() #acceesible through inheritance
lion = Animal()
# lion.species # will not work, as species is not an attribute of the Animal class


#Using the super() Method
# super() method allows a subclass to access methods and attributes of its superclass (reduce redundancy)
# e.g., a Dog class inherits from Animal class
class Animal():
    def __init__(self, species):
        self.species = species

class Dog(Animal):
    def __init__(self, species, name):
        self.name =name
        super(). __init__(species) #declares species attribute defined in Animal 

sam = Dog("canine", "Sammi")
print(sam.species) #accessing the species attribute from the Animal class


#Method Overriding
#Method overriding allows a subclass to provide a specific implementation of a method that is already defined in its superclass
class Animal():
    def makeSound(self):
        print("roar")
class Dog(Animal):
    def makeSound(self):  #overriding the makeSound method
        print("bark")
sam, lion = Dog(), Animal()     #declare multiple variables on a single line

sam.makeSound()
lion.makeSound()

#Inheriting from Multiple Classes
#Python allows a class to inherit from multiple classes
class Physics():
    gravity = 9.8
class Automobile():
    def __init__(self, make, model, year):
        self.make, self.model, self.year = make, model, year #'self' is used to access attributes within the class

class Ford (Physics, Automobile): #able to access Physics and Automobile attributes and methods
    def __init__(self, model, year):
        Automobile. __init__ (self, "Ford", model, year) #super does not work with multiple inheritance

truck = Ford("F-150",2018)
print(truck.gravity, truck.make)    #output both attributes

#HWK
#Create superclass Characters, with attributes name, team, height, weight
class Characters():
    def __init__(self, name, team, height, weight):
        self.name = name
        self.team = team
        self.height = height
        self.weight = weight
    def sayHello(self):
        print("Hello, my name is {} and I'm a {} player ".format(self.name, self.team))


# Subclass based of the superclass, for good team
class GoodPlayers(Characters):
    def __init__(self, name, team, height, weight):
        super().__init__(name, "good", height, weight)  # call the superclass constructor


# Subclass based of the superclass, for bad team
class BadPlayers(Characters):
    def __init__(self, name, team, height, weight):
        super().__init__(name, team, height, weight)  # call the superclass constructor
  
# Instantiate players
player1 = GoodPlayers("Max", "good", 170, 60)
player2 = BadPlayers("Tony", "bad", 180, 75)


# Call the sayHello method for each subclass
player1.sayHello()
player2.sayHello()


        
```
