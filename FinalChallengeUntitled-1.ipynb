```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('data.csv')
print(df.head())

# Perform data analysis
averagetemp = df['Temperature'].mean()
print(f"Average Temperature: {averagetemp:.2f}°C")

# Find maximum temperature
maxtemp = df['Temperature'].max()
print(f"Maximum Temperature: {maxtemp:.2f}°C")

# Find minimum temperature
mintemp = df['Temperature'].min()
print(f"Maximum Temperature: {maxtemp:.2f}°C")

# Plotting line plot showing the temperature over time
plt.plot(df['Date'], df['Temperature'], marker='o', linestyle='-', color='b')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.savefig('temperature_over_time.png')
plt.show()

# Plot histogram displaying the distribution of temperatures
plt.hist(df['Temperature'], bins=20, color='g', alpha=0.7)
plt.title('Temperature Distribution')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('temperature_distribution.png')
plt.show()
# save the histogram pl

# A scatter plot comparing temperature and humidity in two different sets 
plt.scatter(df['Temperature'], df['Humidity'], color='r', alpha=0.5)
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.grid()

# save the scatter plot
plt.tight_layout()
plt.savefig('temperature_vs_humidity.png')
plt.show()










```
